```python
import pandas as pd
import numpy as np

# Load data
df = pd.read_csv('steam-200k.csv')

# use buy behavior
df = df[df['Behavior']==1]

#split data 20,000 purchase
df = df[:20000]
df
```

```python
user = pd.DataFrame()
user['User_ID'] = df['User_ID'].drop_duplicates()
user = user.reset_index(drop=True)
#user.to_csv('Users.csv')
user
```

```python
purchase = pd.DataFrame()
purchase = df[['User_ID','Game','Behavior']]
purchase = purchase.reset_index(drop=True)
#purchase.to_csv('Purchase.csv')
purchase
```

```python
impli_matrix = df_matrix = pd.pivot_table(purchase,index='Game',columns='User_ID',values='Behavior',aggfunc=np.sum,fill_value=0)
impli_matrix
```

```python
corr_matrix=impli_matrix.corr(method='pearson')
corr_matrix
```

```python
###get rid of self correlation
np.fill_diagonal(corr_matrix.values,np.nan)

### sort for top five correlation ( argsort sort min to max so we put minus on the dataset and return index)
order_top5=np.argsort(-corr_matrix.values, axis=1)[:,:5]

### sort for worst correlation
order_worst=np.argsort(corr_matrix.values, axis=1)[:,:1]

### create top5 data frame
result_top5 = pd.DataFrame(
    corr_matrix.columns[order_top5], 
    columns=['1st', '2nd', '3rd', '4th', '5th'],
    index=corr_matrix.index
)
result_top5
```
