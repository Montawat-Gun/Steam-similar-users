```python
import pandas as pd
import numpy as np

# Load data
df = pd.read_csv('steam-200k.csv')

# เอาข้อมูลเพียงแค่ข้อมูลการซื้อ
df = df[df['Behavior']==1]
df = df.drop(columns=['Type'])

#แบ่งข้อมูล
df = df[:20000]
df
```

```python
#สร้างตารางที่หากผู้ใช้ซื้อจะแทนค่าเป็น 1 ผู้ใช้ไม่ได้ซื้อจะเป็น 0
impli_matrix = df_matrix = pd.pivot_table(df,index='Game',columns='User_ID',values='Behavior',aggfunc=np.sum,fill_value=0)
impli_matrix
```

```python
#คำนวณค่า Pearson correlation
corr_matrix=impli_matrix.corr(method='pearson')
corr_matrix
```

```python
#ใส่ User_ID คนที่ต้องการหาว่าใครมีเกมใกล้เคียง 5 อันดับ
user_ID = 1950243
similar_User = pd.DataFrame(corr_matrix.loc[user_ID].sort_values(ascending=False))[1:6]
similar_User
```
